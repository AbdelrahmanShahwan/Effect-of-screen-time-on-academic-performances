```python
import pandas as pd

# Assuming 'df' is your DataFrame

# Iterate through numeric columns and calculate mean, median, and mode
for column_name in df.select_dtypes(include='number').columns:
    mean_value = df[column_name].mean()
    median_value = df[column_name].median()
    mode_value = df[column_name].mode().values  # mode() returns a Series, so use values to get the array

    print(f"\nColumn: {column_name}")
    print(f"Mean: {mean_value}")
    print(f"Median: {median_value}")
    print(f"Mode: {mode_value}")
```

Mapping the hours spent on phone and hours spent studying to make them numerical values

```python

import seaborn as sns
import matplotlib.pyplot as plt
import random

# Assuming 'df' is your DataFrame



# Define a function to generate random numeric values for hours spent on phone
def map_phone(phone_range):
    if '-' in phone_range:
        start, end = map(int, phone_range.split('-'))
        return round(random.uniform(start, end), 2)
    else:
        return float(phone_range)

def map_hours_studying (studying_range):
    if '1-2 Times' in studying_range:
        return random.randint(1, 2)
    elif '2-4 Times' in studying_range:
        return random.randint(2, 4)
    elif '4-6 Times' in studying_range:
        return random.randint(4, 6)
    elif '6-8' in studying_range:
        return random.randint(6, 8)
    elif '10-12' in studying_range:
        return random.randint(10,12)
    else:
        return 0


# Apply the function to map  hours spent on phone randomly within each range
df['How many hours per day do you spend on your phone?'] = df['How many hours per day do you spend on your phone?'].apply(map_phone)
df['On average, how many hours per day do you spend studying or doing homework?'] = df['On average, how many hours per day do you spend studying or doing homework?'].apply(map_hours_studying)

# Check unique values in the column after mapping
print("\nUnique Values After Mapping:")
print("Hours spent on phone:", df['How many hours per day do you spend on your phone?'].unique())
print("Hours studying:", df['On average, how many hours per day do you spend studying or doing homework?'].unique())
```

Line of best fit

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Assuming you have your data in a DataFrame named df
# 'How many hours per day do you spend on your phone?' should be the column name for hours spent on the phone
# 'What is your CGPA?' should be the column name for CGPA

x = df['How many hours per day do you spend on your phone?']  # Assuming this is the column name for hours spent on the phone
y = df['On average, how many hours per day do you spend studying or doing homework?']  # Assuming this is the column name for CGPA

# Convert x and y to numeric values
x = pd.to_numeric(x, errors='coerce')
y = pd.to_numeric(y, errors='coerce')

# Remove any NaN values
x = x.dropna()
y = y.dropna()

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)

# Calculate the predicted values (y_pred) using the equation of the line
y_pred = slope * x + intercept

# Plot the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, label='Data')

# Plot the line of best fit
plt.plot(x, y_pred, color='red', label='Line of Best Fit')

plt.title('Relationship Between Hours Spent on Phone and Hours Studying')
plt.xlabel('Average Hours Spent on Phone')
plt.ylabel('Hours Studying')
plt.legend()
plt.grid(True)
plt.show()
```

Box Plot

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have your data in a DataFrame named df

# Convert the column 'How many hours per day do you spend on your phone?' to numeric values
df['How many hours per day do you spend on your phone?'] = pd.to_numeric(df['How many hours per day do you spend on your phone?'], errors='coerce')

# Define the categories of hours spent on the phone
categories = ['1-2', '2-4', '4-6', '6-8', '10-12']

# Categorize the data based on the ranges of hours spent on the phone
df['Category'] = pd.cut(df['How many hours per day do you spend on your phone?'], bins=len(categories), labels=categories)

# Create a box plot for each category
plt.figure(figsize=(10, 6))
df.boxplot(column='What is your CGPA?', by='Category', ax=plt.gca())
plt.title('Distribution of CGPA Across Different Categories of Hours Spent on Phone')
plt.xlabel('Category of Hours Spent on Phone')
plt.ylabel('CGPA')
plt.suptitle('')
plt.xticks(rotation=45)
plt.show()
```
